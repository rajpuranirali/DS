```python
import seaborn as sns 
```

```python
# Basic Ploting with seaborn 
# inbuit dataset tips
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips
```

```python
#inbuilt database tips
sns.scatterplot(x= 'total_bill' , y = 'tip', data = tips)
plt.title("Scatter Plot of Total Bills vs Tip")
plt.show()
```

```python
# line plot 
sns.lineplot(x = 'size' , y ='total_bill' , data = tips)
plt.title("Line plot of total bill by size")
plt.show()
```

```python
# Categorical Plots
# bar plot 
sns.barplot(x = 'day' , y= 'total_bill' , data = tips)
plt.title('Bar Plot Of Total Bill By Day')
plt.show()
```

```python
# Box Plot
sns.boxplot(x = 'day' , y = 'total_bill' , data = tips)
```

```python
# violin plot
sns.violinplot(x='day', y = 'total_bill' , data = tips)
```

```python
# histogram
sns.histplot(x='day' , y = 'total_bill' , data = tips)
```

```python
# bins=10 means the histogram will divide the range of total_bill into 10 equal-width intervals (or "bins").
# Each bin represents a range of values.
# The height of each bar shows how many data points fall within that bin.
# Example:
# If total_bill ranges from 0 to 50 and bins=10, the bins might look like:
#  0â€“5
# 5â€“10



sns.histplot(tips['total_bill'] , bins = 10 , kde = True)
```

```python
sns.histplot(tips['total_bill'] , bins = 10)
```

```python
# Kde plot 
sns. kdeplot(tips['total_bill'])
```

```python
# Kde plot 
sns. kdeplot(tips['total_bill'], shade= True)
```

```python
# pirplot
sns.pairplot(tips)
```

```python
# HEatmap 
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
# plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x = 'Product_Category' , y = 'Revenue' , data= sales_df)
```

```python
# By default, Seaborn uses estimator=mean, so it would show the average revenue per category.
plt.figure(figsize=(10,6))
sns.barplot(x = 'Product_Category' , y = 'Revenue' , data= sales_df, estimator=sum)
plt.title('Revenue by Product Category')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()
```
