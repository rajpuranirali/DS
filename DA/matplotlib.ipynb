```python
import matplotlib.pyplot as plt
```

```python
# create a line plot
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic line Plot')
```

```python
# customized a line plot 
x = [1,2,3,4,5,9]
y = [1,4,8,10,26,25]
plt.plot(x,y, color='red', linestyle =  '--', marker='o',linewidth =3 ,markersize = '9') #marker = '*'
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic line Plot')
plt.grid(True)
```

```python
# Multiple plots
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2= [1,2,3,4,5]
plt.figure(figsize = (9,5))

plt.subplot(1,2,1) #(1:rows, 2:columns, 1:plot)
plt.plot(x,y1,color = 'green')
plt.title('Plot 1',color = 'blue')
# plt.xlabel('X axis',color = 'blue')
# plt.ylabel('Y axis',color = 'blue')

plt.subplot(1,2,2) #(1:rows, 2:columns, 2:plot)
plt.plot(y1,x,color = 'red')
plt.title('Plot 1',color = 'blue')

```

```python
# bar plot
categories = ['A', 'B' , 'C' , 'D' ,'E']
values = [4 ,25,8,15,25]
plt.bar(categories,values, color = 'purple')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
```

```python
# pie chart
labels = ['A', 'b' , 'c' , 'd']
sizes = [30,20, 40 , 10]
colors = ['gold' ,'yellowgreen' , 'lightcoral' , 'skyblue']


# pie chart 
plt.pie(sizes, labels= labels , colors = colors)
```

```python
# "pull out" one or more slices from the pie 
labels = ['A', 'b' , 'c' , 'd']
sizes = [30,20, 40 , 10]
colors = ['gold' ,'yellowgreen' , 'lightcoral' , 'skyblue']
explode = (0.2,0,0,0)

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
plt.show()
```

```python
# sales data visulization
import pandas as pd
sales_data = pd.read_csv('sales_data.csv')
sales_data.head(5)
```

```python
sales_data.info()

```

```python
# plot total sales by product
total_Sales_by_product = sales_data.groupby('Product_Category')['Revenue'].sum()
total_Sales_by_product
```

```python
total_Sales_by_product.plot(kind='bar', color = 'teal')
```

```python
sub_cat = sales_data.groupby('Sub_Category')['Revenue'].sum()
sub_cat
```

```python
sub_cat.plot(kind='bar', color = 'pink')
```

```python
# sales trend over dare
tred = sales_data.groupby('Year')['Revenue'].sum()
tred
```

```python
tred.plot(kind='bar', color = 'blue')
```
