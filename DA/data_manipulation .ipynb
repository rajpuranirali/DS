```python
# filling missing values with the mean of the column. 

# create a column Price_new named 'Odometer_new'
dd['Odometer_new'] = dd['Odometer (KM)']
dd['Price_new'] = dd['Price']

print(dd.head(20))

```

```python
# print mean of the colun where value is NaN
dd['Odometer_new'] = dd['Odometer (KM)'].fillna(dd['Odometer (KM)'].mean())
print(dd.head(20))
```

```python
# rename the column
dd= dd.rename(columns = {'Price_new' :'Cost'})
dd.head(20)
```

```python
# Change data type from float to int
# it has missing value so it wont convert and will give error. change empty to mean of the coulumn then convert it into int.
dd['doors_new'] = dd['Doors'].astyle(int)

dd.head(20)
```

```python
dd['doors_new'] = dd['Doors'].fillna(dd['Doors'].mean()).astype(int)
dd.head(5)
# dd['new_column_name'] = dd ['existing_clm].astype(datatype)
# dd is readed csv file obj.
```

```python
dd['new_cost'] = dd['Cost'].apply(lambda x:x*2)
dd.head(5)
```
